```python
# Flatten the MultiIndex columns in sp500_data
sp500_data.columns = ['_'.join(col).strip() for col in sp500_data.columns.values]
```

```python
# Ensure Date is a datetime object if it isn't already
sp500_data.index = pd.to_datetime(sp500_data.index)

# Create a 'Year' column from the Date index
sp500_data['Year'] = sp500_data.index.year
```

```python
# Ensure the 'Date' column is in datetime format
Annual_Econ_Data_df['Date'] = pd.to_datetime(Annual_Econ_Data_df['Date'])

# Create the 'Year' column
Annual_Econ_Data_df['Year'] = Annual_Econ_Data_df['Date'].dt.year

# Verify that the 'Year' column has been added
print("Updated Columns in Annual_Econ_Data_df:")
print(Annual_Econ_Data_df.columns)

```

```python
# Perform the merge
merged_df = pd.merge(Annual_Econ_Data_df, sp500_data, on='Year', how='inner')

# Display the merged DataFrame
merged_df.head()
```

```python
# Import necessary library
import pandas as pd


# Group by Year and calculate the mean for the relevant columns
summary_df = merged_df.groupby('Year').agg({
    'Federal Funds Rate': 'mean',  # Average Federal Funds Rate
    'Unemployment Rate': 'mean',    # Average Unemployment Rate
    'GDP': 'mean',                   # Average GDP
    'Inflation Rate': 'mean',        # Average Inflation Rate
    'Adj Close_^GSPC': 'mean'      # Average Adjusted Close for S&P 500
}).reset_index()

# Rename the columns for clarity
summary_df.rename(columns={
    'Adj Close_^GSPC': 'Avg SP500 Adj Close'
}, inplace=True)

# Display the summarized DataFrame
summary_df
```
