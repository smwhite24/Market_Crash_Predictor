```python
import pandas as pd
```

```python
#load the data
data = pd.read_csv('../MARKET_CRASH_PREDICTOR/cleaned_data/final_standardized_data.csv')

#show the first 5 rows of the data
data.head()
```

```python
#Create a pct change colum to track shifts in SP500 column
data['SP500_pct_change'] = data['SP500'].pct_change()
```

```python
#Show the first 5 rows of the data
data.head()
```

```python
#Create a new column called 'crash' based on 10% drops in the percentage change of the SP500
data['crash'] = data['SP500_pct_change'] < -0.1

#Show the first 5 rows of the data
data.head()
```

```python
#Change it from true false to 1 and 0
data['crash'] = data['crash'].astype(int)

#Show the first 5 rows of the data
data.head()
```

```python
#Count the total number of 1s and 0s in the crash column
data['crash'].value_counts()

```
