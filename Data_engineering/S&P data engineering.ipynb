```python
import pandas as pd
import numpy as np
```

```python
#Load the data from the csv file 
data = pd.read_csv('../raw_data/sp500.csv')

#add csv data to a dataframe
df = pd.DataFrame(data)

#Print the first 5 rows of the dataframe
df.head()
```

```python
#make UTC equal True
df['date'] = pd.to_datetime(df['date'], utc=True)

df
```

```python
#Average the open, high, low, and close prices based on the each month
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

df
```

```python
#Mean based on and month
df = df.groupby(['year', 'month']).mean()

#Add to a csv file
df.to_csv('../cleaned_data/sp500_monthly.csv')
```

```python
#Mean based on year
df = df.groupby(['year']).mean()

#Add to a csv file
df.to_csv('../cleaned_data/sp500_yearly.csv')
```

```python
#load the data from the csv file
monthly_data = pd.read_csv('../cleaned_data/sp500_monthly.csv')

#add csv data to a dataframe
monthly_df = pd.DataFrame(monthly_data)

#Print the first 5 rows of the dataframe
monthly_df.head()
```

```python
#Remove the years 1927-1974
monthly_df = monthly_df[monthly_df['year'] > 1974]

#show the first 5 rows of the dataframe
monthly_df.head()
```

```python
#Merge month and year columns under format 'YYYY-MM-DD' with the day set to 1
monthly_df['date'] = pd.to_datetime(monthly_df[['year', 'month']].assign(day=1))

#show the first 5 rows of the dataframe
monthly_df.head()
```

```python
#Drop the year and month columns
monthly_df = monthly_df.drop(columns=['year', 'month'])

#show the first 5 rows of the dataframe
monthly_df.head()
```

```python
#Drop the Open, High, Low, and Volume columns
monthly_df = monthly_df.drop(columns=['Open', 'High', 'Low', 'Volume'])

#show the first 5 rows of the dataframe
monthly_df.head()
```

```python
#create a csv for the monthly_df 
monthly_df.to_csv('../cleaned_data/spstandardized.csv')
```
