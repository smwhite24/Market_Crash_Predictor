```python
import pandas as pd

import yfinance as yf
```

```python
# Fetch historical data for S&P 500 index
sp500 = yf.Ticker("^GSPC")  # S&P 500 index symbol
historical_data = sp500.history(period="max")  # Last 5 years of data
print(historical_data)
```

```python
# Convert to DataFrame (this is already a DataFrame)
sp500_df = historical_data.reset_index()  # Reset index to have 'Date' as a column

# Display the first few rows of the DataFrame
sp500_df
```

```python
# Rename columns for easier access
sp500_df.rename(columns={'Date': 'date'}, inplace=True)

# Check for missing values
print(sp500_df.isnull().sum())

# Display the cleaned DataFrame
sp500_df
```

```python
csv_file_path = "sp500.csv"  # Specify the desired file path
sp500_df.to_csv(csv_file_path, index=False)  # Save DataFrame to CSV without the index

print(f"Data saved to {csv_file_path}")
```
