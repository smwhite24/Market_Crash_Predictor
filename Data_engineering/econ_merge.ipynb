```python
import pandas as pd
```

```python
# Read the CSV file into a DataFrame:

DFF_df = pd.read_csv('../raw_data/Interest_Rate.csv')

DFF_df  # Display the first few rows of the DataFrame
```

```python
DFF_df.rename(columns={'DFF': 'Federal Funds Rate', 'DATE': 'Date'}, inplace=True)

DFF_df.head()
```

```python
# Convert the 'date' column to datetime format
DFF_df['Date'] = pd.to_datetime(DFF_df['Date'])

# Filter the DataFrame to only include data from 1975 onwards
filtered_DFF_df = DFF_df[DFF_df['Date'] >= '1975-01-01']

filtered_DFF_df.head()
```

```python
filtered_DFF_df.to_csv('Cleaned_Interest_Rate.csv', index=False, header=True)

```

```python
# Read the CSV file into a DataFrame:

Inflation_df = pd.read_csv('../raw_data/Inflation.csv')

Inflation_df.head()  # Display the first few rows of the DataFrame
```

```python
Inflation_df.rename(columns={'FPCPITOTLZGUSA': 'Inflation Rate', 'DATE': 'Date'}, inplace=True)

Inflation_df.head()
```

```python
# Convert the 'date' column to datetime format
Inflation_df['Date'] = pd.to_datetime(Inflation_df['Date'])

# Filter the DataFrame to only include data from 1975 onwards
filtered_inflation_df = Inflation_df[Inflation_df['Date'] >= '1975-01-01']

filtered_inflation_df.head()
```

```python
filtered_inflation_df.to_csv('Cleaned_Inflation.csv', index=False, header=True)
```

```python
# Read the CSV file into a DataFrame:

GDP_df = pd.read_csv('../raw_data/GDP.csv')

GDP_df.head()  # Display the first few rows of the DataFrame
```

```python
GDP_df.rename(columns={'DATE': 'Date'}, inplace=True)

GDP_df.head()
```

```python
# Convert the 'date' column to datetime format
GDP_df['Date'] = pd.to_datetime(GDP_df['Date'])

# Filter the DataFrame to only include data from 1975 onwards
filtered_GDP_df = GDP_df[GDP_df['Date'] >= '1975-01-01']

filtered_GDP_df.head()
```

```python
filtered_GDP_df.to_csv('Cleaned_GDP.csv', index=False, header=True)
```

```python
# Read the CSV file into a DataFrame:

unemployment_df = pd.read_csv('../raw_data/Unemployment.csv')

unemployment_df.head()  # Display the first few rows of the DataFrame
```

```python
unemployment_df.rename(columns={'UNRATE': 'Unemployment Rate', 'DATE': 'Date'}, inplace=True)

unemployment_df.head()
```

```python
# Convert the 'date' column to datetime format
unemployment_df['Date'] = pd.to_datetime(unemployment_df['Date'])

# Filter the DataFrame to only include data from 1975 onwards
filtered_UE_df = unemployment_df[unemployment_df['Date'] >= '1975-01-01']

filtered_UE_df.head()
```

```python
filtered_UE_df.to_csv('Cleaned_Unemployment.csv', index=False, header=True)
```

```python
# Merging Interest rate and UE  on the 'date' column
merged_df = pd.merge(filtered_DFF_df, filtered_UE_df, on='Date', how='inner')  # 'inner' for intersection


merged_df.head()
```

```python
# Merge monthly with quarterly
combined_df = pd.merge(merged_df, filtered_GDP_df, on='Date', how='outer')

#Loop through GDP and fill nan values with the previous present value
for i in range(1, len(combined_df)):
	if pd.isnull(combined_df['GDP'][i]):
		combined_df['GDP'][i] = combined_df['GDP'][i-1]

combined_df.head()
```

```python
# Merge the result with yearly
combined_df = pd.merge(combined_df, filtered_inflation_df, on='Date', how='outer')
```

```python
#Remove Inflation Rate_x column
combined_df.drop(['Inflation Rate_x'], axis=1, inplace=True)

#Remove Inflation Rate_y column
combined_df.drop(['Inflation Rate_y'], axis=1, inplace=True)
```

```python
#loop through the combined DataFrame and fill nan values with the previous present value
for i in range(1, len(combined_df)):
	if pd.isnull(combined_df['Inflation Rate'][i]):
		combined_df['Inflation Rate'][i] = combined_df['Inflation Rate'][i-1]

combined_df
```

```python
#Save the cleaned DataFrame to a new CSV file
standardized_df = combined_df.to_csv('standardaized_market_data.csv', index=False, header=True)
```

```python
# Drop rows with NaN values
Annual_Econ_Data_df = combined_df.dropna()

Annual_Econ_Data_df.head()
```
